```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

###1(a) Use the sample mean of each column to fill in the missing attribute.###

# Prepare data
data = pd.read_csv("data/Communities_and_Crime.csv", header=None)
data.drop([0,1,2,3,4], axis=1, inplace=True) # Drop columns that aren't calculable data
data = data.replace("?", np.NaN).astype(np.float64) # Replace all "?" values to be filled
data.fillna(data.mean(), inplace=True) # Fill all nan values with column's mean

###1(d) Turn in the completed data set.###
data.to_csv("data/Communities_and_Crime_Cleaned.csv", index=False, header=False) # Create file
```

```python
###2 Fit the above data using linear regression. Report the 5-fold cross-validation error: The MSE of the best fit achieved on test data, averaged over 5 different 80-20 splits, along with the parameters learnt for each of the five models.###

def getDataMatrices(trainData, testData):
    trainRow = trainData.shape[0]
    trainCol = trainData.shape[1]
    testRow = testData.shape[0]
    testCol = testData.shape[1]
    
    # Get outputs
    trainDataOut = pd.DataFrame([trainData[trainCol-1]])
    testDataOut = pd.DataFrame([testData[testCol-1]])
    # Get w
    w = pd.DataFrame(np.random.randint(low=0, high=10, size=(1, trainCol)))
    
    return trainDataOut, testDataOut, w

def linearRegression(epochs):
    MSEValues, wValues = [], []
    
    for i in range(1, 6):
        trainData = pd.read_csv("data/CandC-train{}.csv".format(i), header=None)
        testData = pd.read_csv("data/CandC-test{}.csv".format(i), header=None)
        trainDataOut, testDataOut, w = getDataMatrices(trainData, testData)
        
        for j in range(epochs):
            outputPredict = np.matmul(w, trainData.T)
            temp = np.subtract(outputPredict, trainDataOut)
            loss = np.matmul(temp, trainData)
            w = w - ((1e-6)*loss)
            
        # Calculate MSE with test data
        outputPredict = np.matmul(w, testData.T)
        temp = np.subtract(outputPredict, testDataOut)
        sqErr = np.power(temp, 2)
        MSE = np.sum(np.sum(sqErr))/sqErr.size
        
        # Update arrays
        MSEValues.append(MSE)
        wValues.append(w.values.tolist()[0])
        
    return MSEValues, wValues


# 5, 80-20 Data Splits
for i in range(1, 6):
    msk = np.random.rand(len(data)) < 0.8
    train = data[msk]
    test = data[~msk]
    # Write to files
    train.to_csv("data/CandC-train{}.csv".format(i), index=False, header=False)
    test.to_csv("data/CandC-test{}.csv".format(i), index=False, header=False)

MSEValues, wValues = linearRegression(1000)
crossValErr = sum(MSEValues)/5
print("5-fold cross-validation error: ", crossValErr, "\n")
for i in range(0, 5):
    print("w values of data set {}: ".format(i+1), wValues[i], "\n")
```

```python
def getWeightedMatrix(X, Y, lambdaVal):
    # X = polynomial matrix
    # Y = output matrix
    # w = weighted matrix = [(X^(T)X+I(lambdaVal))^(-1)] (X^(T)Y)
    XtX_IL_inv = np.linalg.pinv(np.matmul(X.T, X)+(np.identity(X.shape[1])*lambdaVal))
    XtY = np.matmul(X.T, Y.T)
    w = np.matmul(XtX_IL_inv, XtY)
    return w

def getMSE(X, Y, w):
    # MSE(w) = (1/N)sum(from n=1 to N) {Xn^(T)w - Y}^2
    temp = np.power(np.subtract(np.matmul(X, w), Y.T), 2)
    MSE = np.sum(temp)/temp.size
    return MSE

def ridgeRegression():
    lambdaVals = [0.0, 0.1, 0.5, 1.0, 10.0, 100.0, 1000.0]
    MSEValues, wValues = [], []
    lambdaOptimal = 0.0
    MSEOptimal = 99999.0
    
    for i in range(len(lambdaVals)):
        wSetValues = []
        MSESetValue = 0.0
        
        for j in range(1, 6):
            trainData = pd.read_csv("data/CandC-train{}.csv".format(j), header=None)
            testData = pd.read_csv("data/CandC-test{}.csv".format(j), header=None)
            trainDataOut, testDataOut, w = getDataMatrices(trainData, testData)
    
            # Test each data set with each λ value
            w = getWeightedMatrix(trainData, trainDataOut, lambdaVals[i])
            MSE = getMSE(testData, testDataOut, w)
            wSetValues.append(w)
            MSESetValue += MSE
    
        # Keep avg MSE of each data set
        avgMSE = MSESetValue/5
        MSEValues.append(avgMSE)
        # Keep avg w of each data set
        avgW = []
        for k in range(len(wSetValues[1])):
            temp = 0
            for l in range(0, 5):
                temp += wSetValues[l][k]
            avgW.append(temp/5)
        wValues.append(avgW)
        
    MSEOptimal = min(MSEValues, key=float)
    lambdaOptimal = lambdaVals[MSEValues.index(MSEOptimal)]
    
    return MSEValues, wValues, lambdaVals, lambdaOptimal, MSEOptimal

MSEValues, wValues, lambdaVals, lambdaOptimal, MSEOptimal = ridgeRegression()
print("Optimal λ: ", lambdaOptimal)
print("Optimal MSE: ", MSEOptimal, "\n")

for i in range(len(lambdaVals)):
    plt.scatter(lambdaVals[i], MSEValues[i])
    print("MSE value: ", MSEValues[i])
    print("w value: ", wValues[i])
    print("λ value: ", lambdaVals[i], "\n")
    
plt.title("MSE and Lambda values for best fit")
plt.xlabel("Lambda")
plt.ylabel("Avg Test MSE")
plt.show()
```
